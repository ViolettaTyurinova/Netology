```python
# 5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными
# 1 - уволенные
# 0 - работащие
```

```python
satisfaction_level_dism = dataset[dataset['left']==1]['satisfaction_level']
satisfaction_level_work = dataset[dataset['left']==0]['satisfaction_level']

print(stats.ttest_ind(satisfaction_level_dism, satisfaction_level_work))
```

```python
sns.distplot(satisfaction_level_dism, 50)
sns.distplot(satisfaction_level_work, 50)
plt.title('Плотность вероятности распределения уровня удовлетворенности')
plt.legend(['уволенные', 'работающие'])
plt.show()
```

```python
# 10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию
```

####  Гипотеза № 1

```python
H0- уходят из-за несчастного случая на работе
H1- уходят не из-за несчастного случая
```

```python
left_accident = dataset[dataset['Work_accident']==1]['left']
stats.ttest_1samp(left_accident, dataset['left'].mean())
```

```python
так как p-value = 0, то нулевая гипотеза, что сотрудники увльняются изз-за несчастного случая опровергается
```

```python
print(len(dataset[dataset['Work_accident']==1]['left']))
print(len(dataset[dataset['Work_accident']==0]['left']))
```

#### Гипотеза № 2

```python
H0 - стаж работы независит на увольнение
H1 - стаж работы зависит на увольнение
```

```python
time_spend_company = dataset['time_spend_company']
left = dataset['left']
```

```python
from scipy.stats import pearsonr

pearsonr(time_spend_company, left)
```

```python
так как p-value = 0, то гипотеза H0 подтверждается, что значит, стаж работы не влияет на увольнение
```
